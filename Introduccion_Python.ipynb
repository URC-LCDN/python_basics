## ¿Qué es Python?

`Python` es un lenguaje de programación de alto nivel, interpretado y de propósito general que se destaca por su sencillez y legibilidad.

Gracias a su sintaxis sencilla, es un lenguaje recomendado para quienes desean aprender a programar.

Fue creado por Guido van Rossum y lanzado por primera vez en 1991.


El sitio oficial de Python es [python.org](https://www.python.org). En este sitio, puedes encontrar:

- **Descargas**:  
    El intérprete de Python y todos sus módulos, disponibles gratuitamente para las plataformas (MacOS, Windows, Linux, etc).
- **Documentación**:  
    La documentación oficial de Python, que incluye referencias y ejemplos para aprender y utilizar el lenguaje.
- **Tutoriales**:  
    Tutoriales para ayudar a desarrolladores principiantes como avanzados a aprender y mejorar sus habilidades en Python.
- **Herramientas**:  
    Recursos y herramientas adicionales para el desarrollo en Python, como entornos de desarrollo integrado (IDE) y bibliotecas adicionales.
- **Comunidad**:  
    Información sobre la comunidad de Python, incluyendo foros y eventos relacionados con `Python` como la [PyCon](https://pycon.org/) o [PyConLatam](https://www.pylatam.org/).


El logo de Python, que representa dos serpientes, es un juego visual basado en el nombre del lenguaje. Aunque el nombre `Python` en realidad proviene del grupo de comedia británico __Monty Python__ y no de la serpiente, la imagen de la serpiente encajó bien y se convirtió en un símbolo visualmente reconocible y fácil de asociar con el lenguaje de programación.

### Versiones de Python

Python ha tenido tres versiones a lo largo de los años, cada una con sus propias características y mejoras.

 - **Python 1.x**  
    Lanzada en enero de 1994, esta fue la primera versión oficial de Python. Estableció muchas de las características básicas del lenguaje que aún se utilizan hoy en día.

 - **Python 2.x**  
    Lanzada en octubre de 2000, introdujó muchas mejoras, incluyendo la las listas por comprensión. Python 2.7, lanzada en julio de 2010, fue la última versión de la serie 2.x. El soporte oficial para dicha versión  finalizó el 1 de enero de 2020.
    
    Desde entonces, se recomienda a todos los usuarios migrar a Python 3.

 - **Python 3.x**.
    Lanzada en diciembre de 2008, una de las mejoras que trajó esta versión fue la introducción de los f-strings para la formateo de cadenas de manera más sencilla.  
    
    La última versión estable es Python 3.12 y fue, lanzada en octubre de 2023, incluye mejoras continuas en el rendimiento principalmente.


Actualmente `Python3` se encuentra incorporado nativamente en Linux y MacOs.  


Se puede consultar el estatus de las versiones de `Python` en [https://devguide.python.org/versions/](https://devguide.python.org/versions/).

### Ventajas y desventajas de usar Python

#### Ventajas

- **Lenguaje de alto nivel.**   
    Para programar con Python se pueden usar elementos del lenguaje natural, ya que tiene una sintaxis similar al inglés, por lo que es fácil de leer, escribir y aprender.

- **Polivalente y de paradigmas.**  
    Al ser un lenguaje de propósito general, se puede usar para diversos propósitos.

- **Módulos**.  
    Tiene una amplia colección de módulos/frameworks.

- **Portabilidad.**  
    Es compatible con todos los sistemas operativos (macOS, Linux, y Windows).

- **Gratis.**  
    Python es un lenguaje de programación desarrollado bajo la licencia de código abierto

- **Baja curva de aprendizaje.**  
    La sencillez de la sintaxis de Python permite escribir programas totalmente funcionales en pocas líneas de código, por lo que su curva de aprendizaje es muy baja.

- **Comunidad.**  
    El hecho de que Python sea gratuito y de código abierto contribuye a crear una comunidad sólida

#### Deventajas

- **Lentitud.**  
    La lentitud de Python se debe principalmente a su naturaleza dinámica y versatilidad.

- **Consumo de memoria.**  
    El consumo de memoria de Python es muy alto, y esto se debe a la flexibilidad de los tipos de datos.

- **Desarrollo móvil.**  
    Python es ideal para plataformas de escritorio y servidor, pero para el desarrollo móvil no es un lenguaje muy adecuado.

### Popularidad de Python

La popularidad de `Python` se ha disparado en los últimos años y se puede atribuir a varios factores:

* **Versatilidad y Aplicaciones Amplias**.
    Python se usa en una variedad de campos, incluyendo:
    * _Desarrollo Web:_  
    Con frameworks como Django y Flask.
    * _Ciencia de Datos y Análisis:_  
    Con bibliotecas como pandas, NumPy y scikit-learn.
    * _Inteligencia Artificial y Machine Learning:_  
    Utilizando TensorFlow, PyTorch y otras herramientas.

    Para tareas de automatización y desarrollo de scripts.

* **Popularidad en la Industria**.  
    Es ampliamente adoptado en la industria tecnológica y en empresas de todos tamaños. Su uso se extiende desde startups hasta grandes corporaciones, lo que refuerza su relevancia y demanda en el mercado laboral.

### `Python` para Ciencia de Datos

`Python` se ha consolidado como una herramienta fundamental en el campo de la ciencia de datos debido a su rica colección de módulos y su versatilidad.


**Módulos Especializadas**

- **NumPy**:  
    Proporciona soporte para realizar operaciones de grandes matrices. Es fundamental para realizar cálculos numéricos y operaciones algebraicas, formando la base para muchas otras bibliotecas en Python.

- **Pandas**:  
    Facilita la manipulación y análisis de datos estructurados. Con estructuras de datos como DataFrames, pandas permite realizar operaciones complejas de filtrado, agregación y transformación de datos de manera eficiente.

- **TensorFlow**:   
    Módulo para el desarrollo y entrenamiento de modelos de machine learning y deep learning. TensorFlow ofrece una plataforma robusta para construir y desplegar redes neuronales complejas.

- **PyTorch**:   
    Otra popular biblioteca de deep learning que proporciona una interfaz flexible y dinámica para construir y entrenar modelos.

- **Scikit-Learn**:  
    Ofrece herramientas simples y eficientes para el análisis de datos y el modelado predictivo. Incluye algoritmos para clasificación, regresión, clustering, y preprocesamiento de datos, facilitando la implementación de técnicas de machine learning.

- **Plotly**:   
    Una biblioteca para la creación de visualizaciones interactivas y atractivas. Plotly permite la creación de gráficos y dashboards que son útiles para explorar y comunicar datos de manera efectiva.

- **nltk (Natural Language Toolkit)**:  
    Proporciona herramientas y recursos para el procesamiento de lenguaje natural (NLP). NLTK es útil para tareas como tokenización, etiquetado de partes del discurso y análisis de sentimientos.
