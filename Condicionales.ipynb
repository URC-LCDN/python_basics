# **Estructuras de control**

Las estructuras de control, denominadas también sentencias de control, permiten tomar decisiones y realizar un proceso repetidas veces. Son el conjunto de reglas que permiten controlar el flujo de ejecución de las instrucciones de un algoritmo o de un programa.

La mayoría de los lenguajes de programación actuales soportan o utilizan las mismas estructuras de control o, al menos, son muy parecidas.

Los condicionales `if`, `elif`, `else` te permiten controlar el rumbo de la ejecución de un programa.

image.png

### Explicación:

1. **Condicional `if`**:
Se usa para ejecutar un bloque de código solo si una condición es verdadera.

```
edad = input("Tu edad es: ")
edad = int(edad)
```

```
if edad < 18:
    print("Eres menor de edad.")
```

2. **Condicional `elif`**:
   Significa "else if". Se usa para verificar múltiples condiciones. Solo se ejecuta si la condición anterior en el bloque `if` es falsa.

```
if edad < 18:
    print("Eres menor de edad.")
elif edad >= 18 and edad < 65:
      print("Tienes edad para votar.")
```

3. **Condicional `else`**:
Se usa como un bloque final que se ejecuta si todas las condiciones anteriores son falsas.

```
if edad < 18:
      print("Eres menor de edad.")
elif edad >= 18 and edad < 65:
      print("Tienes edad para votar.")
else:
      print("Eres un adulto mayor.")
```

Observación:

Cuando necesitamos comprobar más de una condicion utilizamos `elif` tantas veces como sea necesario. También, mediante un `else` podemos destinar la ejecucicon de instrucciones si no se satisface ninguna condicion impuesta.

Otro ejemplo del uso del `if-else`, utilizando la función `input`

```
x = int(input("Introduce un numero entero: "))
if x < 0:
  x = 0
  print('Si es negativo lo cambio a cero')
elif x == 0:
  print('Cero')
elif x == 1:
  print('Uno')
else:
  print('Mayor a 1')
```
